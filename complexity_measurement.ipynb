```python
# -------------------------------------------
# CALCULATE AND RECORD LORENA COMPLEXITY MEASURES
# -------------------------------------------
# Loop through and calculate the complexity of datasets selected in the previous code cell.
# Calculate Lorena's 22 complexity measures plus the overall score for each dataset.
# Capture additional dataset metadata such as number of samples, features, and class imbalance.
# Save results to an Excel file, appended to any existing results if the file already exists.

import pandas as pd
import numpy as np
import problexity as px
from pathlib import Path
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import time
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

MAX_FEATURES = 30    # To control runtime, randomly select up to this many features from each dataset before running problexity.

# ── Path definitions ──────────────────────────────────────────────────────────
# INSTRUCTIONS: Update these paths to point to your local folders and files.
# This notebook is designed to run locally on a CPU-equipped machine, not Colab.

USER_PATH     = Path(r"C:\Your\Local\Project\Folder")

DATASETS_PATH = USER_PATH / "datasets"
MANIFEST_PATH = USER_PATH / "dataset_manifest_final.xlsx"
ERRORS_PATH   = USER_PATH / "benchmark_problexity_errors.csv"

# ── Configuration ─────────────────────────────────────────────────────────────
MAX_ROWS = 15000     # Problexity runtime was 1 to 5 hours on six of 182 datasets, less than 1 minute on 113, the rest no more than 15 minutes.

# ══ Main manifest loop ════════════════════════════════════════════════════════
results = []

for i, dataset_file in enumerate(DATASETS, start=1):    # DATASETS contains those selected in the previous code cell.
    dataset_path = DATASETS_PATH / dataset_file
    print(f"\n{'='*70}")
    print(f"Processing: {dataset_file} ({i} of {num_datasets})")
    print(f"{'='*70}")
    dataset_start = time.time()

    try:
        # ── Load dataset ──────────────────────────────────────────────────────
        df = pd.read_csv(dataset_path)
        print(f"  Shape: {df.shape[0]:,.0f} rows × {df.shape[1]:,.0f} columns")

        # ── Full dataset metrics ──────────────────────────────────────────────
        n_samples  = len(df)
        X_full     = df.drop(columns=['target'])
        y_full     = df['target']

        cat_cols   = [col for col in X_full.columns if X_full[col].dtype == 'object']
        num_cols   = [col for col in X_full.columns if X_full[col].dtype != 'object']

        n_features             = X_full.shape[1]
        n_numeric_features     = len(num_cols)
        n_categorical_features = len(cat_cols)

        # ── Max unique categorical column BEFORE encoding ─────────────────────
        if cat_cols:
            unique_counts  = X_full[cat_cols].nunique()
            max_unique_col = unique_counts.idxmax()
            max_unique_val = int(unique_counts.max())
        else:
            max_unique_col = None
            max_unique_val = 0

        # ── Class imbalance ratio ─────────────────────────────────────────────
        class_counts    = y_full.value_counts()
        class_imbalance = round(class_counts.min() / class_counts.max(), 4)

        # ── Randomly select up to MAX_FEATURES features ───────────────────────
        if X_full.shape[1] > MAX_FEATURES:
            X_full = X_full.sample(n=MAX_FEATURES, axis=1, random_state=42)
            print(f"  ✂️ Features randomly reduced from {n_features} to {MAX_FEATURES}")
            # Recompute column types after reduction
            cat_cols = [col for col in X_full.columns if X_full[col].dtype == 'object']
            num_cols = [col for col in X_full.columns if X_full[col].dtype != 'object']

        # ── Preprocessing ─────────────────────────────────────────────────────
        df_clean = pd.concat([X_full, y_full], axis=1).dropna()
        X_clean  = df_clean.drop(columns=['target'])
        y_clean  = df_clean['target']

        # Ordinal encode categoricals
        if cat_cols:
            enc = OrdinalEncoder()
            X_clean[cat_cols] = enc.fit_transform(X_clean[cat_cols])

        # StandardScaler on numerics
        scaler = StandardScaler()
        X_clean[num_cols] = scaler.fit_transform(X_clean[num_cols])

        # ── Stratified sample to MAX_ROWS ─────────────────────────────────────
        df_scaled = pd.concat([X_clean, y_clean], axis=1)
        if len(df_scaled) > MAX_ROWS:
            df_scaled, _ = train_test_split(
                df_scaled,
                train_size=MAX_ROWS,
                stratify=df_scaled['target'],        # Preserve the class distribution when sampling rows.
                random_state=42
            )
            df_scaled = df_scaled.reset_index(drop=True)
        
        # Confirm sample shape and class distribution
        print(f"  📐 Sample shape: {df_scaled.shape[0]:,.0f} rows × {df_scaled.shape[1]-1:,.0f} features")
        
        # ── Run problexity ────────────────────────────────────────────────────
        cc = px.ComplexityCalculator()
        cc.fit(df_scaled.drop(columns=['target']), df_scaled['target'])
        report = cc.report()

        # ── Assemble results row ──────────────────────────────────────────────
        runtime_seconds = round(time.time() - dataset_start, 2)

        row = {
            'dataset':                dataset_file,
            'n_samples':              n_samples,
            'n_features':             n_features,
            'n_numeric_features':     n_numeric_features,
            'n_categorical_features': n_categorical_features,
            'max_unique_values':      max_unique_val,
            'max_unique_column':      max_unique_col,
            'class_imbalance':        class_imbalance,
            'score':                  report['score'],
        }
        row.update(report['complexities'])
        row['runtime_seconds'] = runtime_seconds

        results.append(row)

        print(f"  ✅ Completed in {runtime_seconds:.1f}s | Score: {report['score']:.4f} | Imbalance: {class_imbalance:.4f}")

    except Exception as e:
        print(f"  ❌ FAILED: {e}")
        save_error(dataset_file, 'MANIFEST', e)

# ══ Write Excel manifest ══════════════════════════════════════════════════════
if results:
    manifest_df = pd.DataFrame(results)
    if MANIFEST_PATH.exists():
        existing_df = pd.read_excel(MANIFEST_PATH, sheet_name='Benchmark Manifest')
        manifest_df = pd.concat([existing_df, manifest_df], ignore_index=True)
    with pd.ExcelWriter(MANIFEST_PATH, engine='openpyxl') as writer:
        manifest_df.to_excel(writer, sheet_name='Benchmark Manifest', index=False)
    print(f"\n{'='*60}")
    print(f"Manifest saved: {MANIFEST_PATH}")
    print(f"Datasets processed: {len(results)}")
    print(f"{'='*60}")
else:
    print("No results to save.")
```
